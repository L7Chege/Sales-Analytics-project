Additionally, I can introduce <u><mark>world population data</mark></u> since if the mortality rate is high, the population growth rate becomes negative. Hence, as the mortality rate increases, the population growth decreases.
Population change is governed by the balance between birth rates and death rates. If the birth rate stays the same and the death rate decreases, then population numbers will grow. If the birth rate increases and the death rate stays the same, then population will also grow.
People living in a low-income country are far more likely to die of a communicable disease than a non-communicable disease. Despite the global decline, six of the top 10 causes of death in low-income countries are communicable diseases. Malaria, tuberculosis and HIV/AIDS all remain in the top 10.
