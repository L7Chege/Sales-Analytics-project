# Data Visualization

1. Scatter plot of sales vs profit

```python
sns.regplot(x='Sales', y='Profit', data=df, scatter_kws={"color": "green"}, line_kws={"color":"black"})
```

## Finding a correlation between profit and sales

```python
plt.rcParams['figure.figsize'] = (15,6)
```

```python
correlation_matrix = df.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('sales')
plt.ylabel('profit')
```

A correlation coefficient of 0.48 would suggest a moderate positive relationship between the two variables.

- Central has 23.2%: This suggests that the Central region has the third-highest total sales among all regions in the USA. Possible reasons for this could include a strong manufacturing and agricultural industry in the region, as well as the presence of major cities such as Chicago and Dallas.


- South has 16.2%: This indicates that the South region has the lowest total sales among all regions in the USA. Possible reasons for this could include lower population density in the region, a more rural economy, and lower levels of urbanization compared to other regions.

```python
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

top_states_p = df.groupby("State").Profit.sum().nlargest(n=10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar' ,figsize =(10,6), fontsize=13)
top_states_p.plot(kind = 'bar' ,figsize =(10,6), fontsize=13, color='red')
plt.xlabel('States')
plt.ylabel('Total Sales')
plt.title('Top State earners in Sales vs Proft', fontsize = 20)
```

From the above graph, most sales are majorly concentrated in states: California, New York and Washington =>150,000 sales.

- Why are those the best performing states?

As of the 2020 United States Census, the population of <b>California is approximately 39.5 million</b>, the population of <b>New York is approximately 20.2 million</b>, and the population of <b>Washington is approximately 7.8 million.</b>

Urban centers: California, New York, and Washington are home to several large urban centers, such as Los Angeles, San Francisco, New York City, and Seattle, which are hubs for commerce, tourism, and business.


```python
least_states_s = df.groupby("State").Sales.sum().nsmallest(n=10)

least_states_p = df.groupby("State").Profit.sum().nsmallest(n=10)

plt.style.use('seaborn')
least_states_s.plot(kind = 'bar' ,figsize = (10,6), fontsize = 14)
least_states_p.plot(kind = 'bar' ,figsize = (10,6), fontsize = 14, color = 'red')
plt.xlabel('States')
plt.ylabel('Total Sales')
plt.title('Bottom 10 States Sales vs Profit', fontsize = 20)
plt.show()
```

There could be several reasons why the business might be making losses in Texas, Ohio, and Pennsylvania despite having good sales. Here are a few possible reasons:

- High operating costs: Texas, Ohio, and Pennsylvania have relatively high operating costs compared to some other states, including costs related to labor, rent, and taxes. This can put a strain on businesses, especially those with slim profit margins.


- Intense competition: These states have a high number of businesses and can be highly competitive markets. If a business is unable to differentiate itself or offer competitive pricing or value, it may struggle to remain profitable despite having good sales.


- Economic conditions: Despite having good sales, the economic conditions of these states may not be favorable for some businesses. For example, if there is high unemployment or low consumer confidence, it can impact the spending habits of consumers and make it difficult for businesses to thrive.


- Supply chain disruptions: With the global supply chain being impacted by various factors, such as COVID-19 pandemic, businesses in Texas, Ohio, and Pennsylvania may face supply chain disruptions and increased costs associated with acquiring goods and raw materials.


- Regulatory environment: Businesses in these states may face more regulations or bureaucratic hurdles that make it difficult to operate or achieve profitability, especially if they are in industries that are highly regulated.
